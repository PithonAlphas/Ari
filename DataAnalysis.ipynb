## Scatter plot

```python
figure, axs = plt.subplots(1,2, sharey = True, sharex = False, squeeze = False)
data.plot(kind="scatter", x = "Escuelas", y = "Tasa de delitos por 10 mil habitantes", ax = axs[0][0])
data.plot(kind="scatter", x = "Población", y = "Tasa de delitos por 10 mil habitantes", ax = axs[0][1])
```

```python
figure, axs = plt.subplots(1,2, sharey = True, sharex = False, squeeze = False)
data.plot(kind="scatter", x = "Número de policías", y = "Tasa de delitos por 10 mil habitantes", ax = axs[0][0])
data.plot(kind="scatter", x = "Tasa de policías por habitantes", y = "Tasa de delitos por 10 mil habitantes", ax = axs[0][1])
```

## Histograma

```python
plt.hist(data["Escuelas"], bins = int(np.ceil(1+np.log2(3333))))
plt.xlabel("Número de Escuelas")
plt.ylabel("Frecuencia")
plt.title("Histograma de escuelas")
```

## Boxplot, diagrama de caja y bigotes

```python
plt.boxplot(data["Escuelas"])
plt.ylabel("Número de escuelas")
plt.title("Boxplot de las escuelas")
```
